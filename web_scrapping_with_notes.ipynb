```python
# Importing the Libraries for web scrapping
from bs4 import BeautifulSoup
import requests
```

```python
# Define the url
url = requests.get('https://www.meesho.com/tops-ladies/pl/3ja')
```

Also check for the status code. If status code is 200 then only proceed to extract the data otherwise not.

You can proceed with parsing the content.

*Parsing* = Parsing, syntax analysis, or syntactic analysis is a process of analyzing a string of symbols, either in natural language, computer languages or data structures, conforming to the rules of a formal grammar by breaking it into parts.

```python
soup = BeautifulSoup(url1.text, "html.parser")
```

The above step is to parse the HTML content using BeautifulSoup.

```python
print(soup.prettify())
```

Extract the title of the wikipedia page

```python
title = soup.find('h1').text
print(title)
```

To extract all the paragraph from the page

```python
paragraph = soup.find_all ('p')
print("\nExtracted Pragraph(First 3):")
for i, para in enumerate(paragraph[:3]): # To display only first 3 paragraphs
    print(f"\nParagraph {i+1}: {para.text.strip()}")
```

Extracting all the hyperlinks on the page

```python
link = soup.find_all("a", href=True) # Find all links within "a" tag
print("\nExtracted Hyperlinks(First 5):")
for i, link in enumerate(link[:5]): # Display First 5 links
    print(f"Link{i+1}:{link['href']}")
```

```python
table_header = soup.find_all('th') # to find the header of the table 
print(table_header)
```

```python
t_header = [title.text.strip() for title in table_header]
print(t_header)
```
